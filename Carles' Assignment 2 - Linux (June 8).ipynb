In NumPy:
- dimensions: are called axes
- the number of axes: is a rank

For example:
the coordinates of a POINT in 3D space [1, 2, 1] is an array of rank 1, because it has one axis.
That axis has a length of 3.

The array has rank 2 (it is 2-dimensional)
- The first dimension (axis) has a length of 2
- The second dimension has a length of 3
